```python
import os
train_img_path = "/content/data2/images/train"
val_img_path = "/content/data2/images/val"
```

```python
#Training images
with open('/content/data2/train.txt', "a+") as f:
  img_list = os.listdir(train_img_path)
  for img in img_list:
    f.write(os.path.join(train_img_path,img+'\n'))
  print("Done")
```

```python
# Validation Image
with open('/content/data2/val.txt', "a+") as f:
  img_list = os.listdir(val_img_path)
  for img in img_list:
    f.write(os.path.join(val_img_path,img+'\n'))
  print("Done")
```
